```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:,.10f}'.format
shib = pd.read_csv("SHIB-USD.csv")
shib.tail()
```

```python
# plot the trend 
plt.style.use("ggplot")
shib.set_index('Date').plot(figsize = (20, 10))
plt.show()
```
